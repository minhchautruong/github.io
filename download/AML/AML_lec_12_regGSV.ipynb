```python
# import all packages needed
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Read and prepare dataset
df = pd.read_csv("housing_price.csv", header = 0, na_values ="NA",  comment ='\t', sep=',', skipinitialspace = True)
df.info()
```

```python
# Creat the feature input X and target variable Y
X = df.iloc[:,:12]
Y = df.iloc[:,12]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.20,random_state=42)
```

```python
# import the decision tree regression 
from sklearn import tree 
 
#import all modules and functions
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV, train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python

knn_model = KNeighborsRegressor()
decision_tree_model = DecisionTreeRegressor()

# Define parameter grids for both models
knn_param_grid = {
    'n_neighbors': [3, 5, 7, 9],
    'weights': ['uniform', 'distance']
}

decision_tree_param_grid = {
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10]
}

# Create GridSearchCV objects for both models
knn_grid_search = GridSearchCV(estimator=knn_model, param_grid=knn_param_grid, scoring='neg_mean_absolute_error', cv=5)
decision_tree_grid_search = GridSearchCV(estimator=decision_tree_model, param_grid=decision_tree_param_grid, scoring='neg_mean_absolute_error', cv=5)

# Fit the grid searches to the training data
knn_grid_search.fit(X_train, y_train)
decision_tree_grid_search.fit(X_train, y_train)

# Print the best hyperparameters and their corresponding mean squared error scores for both models
print("Best hyperparameters for KNN:", knn_grid_search.best_params_)
print("Best negative mean squared error for KNN:", knn_grid_search.best_score_)

print("Best hyperparameters for Decision Tree:", decision_tree_grid_search.best_params_)
print("Best negative mean squared error for Decision Tree:", decision_tree_grid_search.best_score_)

# Evaluate the best models on the test data
knn_best_model = knn_grid_search.best_estimator_
decision_tree_best_model = decision_tree_grid_search.best_estimator_

knn_test_score = knn_best_model.score(X_test, y_test)
decision_tree_test_score = decision_tree_best_model.score(X_test, y_test)

print("Test mean absolute error for best KNN model:", knn_test_score)
print("Test mean absolute error for best Decision Tree model:", decision_tree_test_score)
```

```python
# scatter plot for original data
y_pred_dt = decision_tree_best_model.predict(X_test)
plt.scatter(y_test, y_pred_dt)
plt.xlabel("True prices from test set")
plt.ylabel("Predicted prices from trained model")
plt.show()
```
