```python
y = np.array([40, 20, 25, 15])
mylabels = ["orange", "Banana", "Kiwi", "Apple"]

myexplode = [0, 0 , 0 , 0.1]
plt.pie(y,  labels = mylabels, explode = myexplode, shadow = True)

plt.title("This is a fruit distribution")
plt.show()
```

```python
fig, axes = plt.subplots(1,2)

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y1 = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
y2 = 300 -y1
y3 = y1 - 100


axes[0].grid()
axes[0].plot(x,y1,"*-r", label="Plot 1")
axes[0].legend(loc = "upper left")

axes[1].grid()
axes[1].plot(x, y2,"*-r", label ="Plot 2")

axes[1].plot(x, y3,"+--b", label ="Plot 3")
axes[1].legend(loc = "upper left")
plt.show()
```
